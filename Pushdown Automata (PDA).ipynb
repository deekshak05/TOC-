```python
PDA accepting strings of type a^nb^n where n>=1
```

```python
#testcase1_acceptableString
pda=dict()

pda[(0,'a')]=1
pda[(0,'b')]=4
pda[(0,'z')]=4
pda[(1,'a')]=1
pda[(1,'b')]=2
pda[(2,'a')]=4
pda[(2,'b')]=2
pda[(2,'z')]=3
pda[(4,'a')]=4
pda[(4,'b')]=4
pda[(4,'z')]=4

class stack:
    def __init__ (self,pda):
        self.items=[]
    def push(self,item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def isempty(self):
        return self.items==[]
    def size(self):
        return len(self.items)
s=stack(pda)

    
x=input("Enter the string to be checked : ")
s.push('z')
cs=0
fs=3
for i in x:
    top=s.pop()
    
    s.push(top)
    cs=pda[cs,i]
    print("current state is :",cs)
    if i=='a' and top=='z':
        s.push(i)
        
      
    elif i=='a' and top=='a':
        s.push(i)
       
    elif i=='b' and top=='a':
        s.pop()
    elif i=='b' and top!='a':
        s.push(i)
        
y=s.pop()
cs=pda[cs,y]
print("current state is :",cs)


if cs==fs:
    print("final state")
    print("STRING ACCEPTED !")
else:
    print("stuck")
    print("STRING NOT ACCEPTED !")

```
